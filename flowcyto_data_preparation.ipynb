# Preparation of the FlowCytometry data:

```python
import pandas as pd
import numpy as np
```

### Load in CSV files of the different cell groups obtained from manual gating with FlowCytometry analysis software:

```python
#total combined file:
total_df_ungrouped = pd.read_csv("data/export_compensated_total.csv")

#Total live cells:
live_df_ungrouped = pd.read_csv("data/export_compensated_total_live.csv")

#CD4 subtypes:
CD4_tcm_df_ungrouped = pd.read_csv("data/export_compensated_total_CD4_tcm.csv")
CD4_tem_df_ungrouped = pd.read_csv("data/export_compensated_total_CD4_tem.csv")
CD4_temra_df_ungrouped = pd.read_csv("data/export_compensated_total_CD4_temra.csv")
CD4_th0_df_ungrouped = pd.read_csv("data/export_compensated_total_CD4_th0.csv")

#CD8 subtypes:
CD8_tcm_df_ungrouped = pd.read_csv("data/export_compensated_total_CD8_tcm.csv")
CD8_tem_df_ungrouped = pd.read_csv("data/export_compensated_total_CD8_tem.csv")
CD8_temra_df_ungrouped = pd.read_csv("data/export_compensated_total_CD8_temra.csv")
CD8_tc0_df_ungrouped = pd.read_csv("data/export_compensated_total_CD8_tc0.csv")



total_df_ungrouped.head()
```

```python
print("total",total_df_ungrouped.shape)
print("live", live_df_ungrouped.shape)

print("CD4_tcm", CD4_tcm_df_ungrouped.shape)
print("CD4_tem", CD4_tem_df_ungrouped.shape)
print("CD4_temra", CD4_temra_df_ungrouped.shape)
print("CD4_th0", CD4_th0_df_ungrouped.shape)
print("CD8_tcm", CD8_tcm_df_ungrouped.shape)
print("CD8_tem", CD8_tem_df_ungrouped.shape)
print("CD8_temra", CD8_temra_df_ungrouped.shape)
print("CD8_tc0", CD8_tc0_df_ungrouped.shape)
```

### Create a dataframe with none CD4/CD8 live cells:

```python
#Remove none CD4/CD8 by concat and removing duplicates from the total live dataframe:
onlylive_df = pd.concat([live_df_ungrouped, CD4_tcm_df_ungrouped, CD4_tem_df_ungrouped, CD4_temra_df_ungrouped,
                    CD4_th0_df_ungrouped, CD8_tcm_df_ungrouped, CD8_tem_df_ungrouped,
                    CD8_temra_df_ungrouped, CD8_tc0_df_ungrouped]).drop_duplicates(keep=False).reset_index(drop=True)
onlylive_df.shape
```

```python
#Validate if the correct number of cells was substracted from the live dataframe by comparing shape:
4767912-(472548+471606+129912+913463+27806+261363+254410+323249)
```

### Create dataframe containing double cells and debris:

```python
rest_df = pd.concat([total_df_ungrouped, live_df_ungrouped]).drop_duplicates(keep=False).reset_index(drop=True)
```

```python
rest_df.to_csv("data/dead_debris_ungrouped.csv", index = False)
```

```python
rest_df_grouped = rest_df.copy()
rest_df_grouped["class"] = "debris"
rest_df_grouped.to_csv("data/rest_df_grouped.csv", index = False)
rest_df_grouped.head()
```

```python
onlylive_df.to_csv("data/onlylive_ungrouped.csv", index = False)
```

```python
cells_df = pd.concat([CD4_tcm_df_grouped, CD4_tem_df_grouped, CD4_temra_df_grouped,
                    CD4_th0_df_grouped, CD8_tcm_df_grouped, CD8_tem_df_grouped,
                    CD8_temra_df_grouped, CD8_tc0_df_grouped])
cells_df
```
